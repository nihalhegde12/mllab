```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dataset/marklist.csv')
print(df.head())
```

```python
col1="Internal"
col2="External"
selected_data=df[[col1,col2]]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(selected_data[col1],selected_data[col2],color='blue',alpha=0.8)
plt.title(f' scatter plot of {col1} vs {col2}')
plt.xlabel(col1)
plt.ylabel(col2)
plt.grid(True)
plt.show()
```

```python
cor_coef=np.corrcoef(selected_data[col1],selected_data[col2])[0,1]
print(f'pearson correlation coefficient ({col1} vs {col2}): {cor_coef}')
```

```python
cov_matrix=np.cov(selected_data[col1],selected_data[col2])
print("covariance matrix:")
print(cov_matrix)
```

```python
correlation_matrix = selected_data.corr()
print("\ncorrelation matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("heatmap")
plt.show()
```
